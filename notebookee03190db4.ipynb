```python
import pandas as pd
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Load data 
train = pd.read_csv('/kaggle/input/titanic/train.csv')
test = pd.read_csv('/kaggle/input/titanic/test.csv')
```

```python
train.head()
```

```python
test.head()
```

```python
# TAHAP 1: PEMBERSIHAN DATA (DATA CLEANING)

# 1. Mengisi data umur (Age) yang kosong dengan nilai tengah (median)
train['Age'] = train['Age'].fillna(train['Age'].median())
test['Age'] = test['Age'].fillna(test['Age'].median())

# 2. Mengisi data tarif (Fare) yang kosong di data test
test['Fare'] = test['Fare'].fillna(test['Fare'].median())

# 3. Mengubah data teks 'Sex' menjadi angka agar bisa dibaca mesin
# male = 0, female = 1
train['Sex'] = train['Sex'].map({'male': 0, 'female': 1})
test['Sex'] = test['Sex'].map({'male': 0, 'female': 1})
```

```python
# TAHAP 2: MEMILIH FITUR (FEATURE SELECTION)

# Kita pilih kolom yang paling berpengaruh: Kelas, Jenis Kelamin, Umur, Saudara, Orangtua, Tarif
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']

X = train[features]           # Data untuk belajar
y = train['Survived']         # Jawaban kunci (Target)
X_test = test[features]       # Data soal ujian (yang mau diprediksi)
```

```python


# TAHAP 3: MELATIH MODEL (TRAINING)

# Kita menggunakan Random Forest karena biasanya akurat untuk data Titanic
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)

# TAHAP 4: PREDIKSI & SIMPAN HASIL

predictions = model.predict(X_test)

# Membuat DataFrame untuk hasil akhir
output = pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': predictions})

# Menyimpan ke file CSV (siap untuk submit ke Kaggle)
output.to_csv('submission.csv', index=False)

print("Sukses! File 'submission.csv' berhasil dibuat.")
print(output.head()) # Menampilkan 5 baris pertama hasil prediksi
```
